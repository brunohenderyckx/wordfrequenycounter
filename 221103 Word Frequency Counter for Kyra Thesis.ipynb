```python
import docx
from collections import Counter
import pandas as pd
```

```python
# open connection to Word Document
doc = docx.Document(r"C:\Users\bruno.henderyckx\Desktop\Thesis\Thesis - v9 final draft.docx")
 
# read in each paragraph in file
result = [p.text for p in doc.paragraphs]
print(result[0:500])
```

```python
results_string = "".join(result)

print(results_string[0:500])
      
print(results_string[0:1500].replace('\n'," ").split(" "))
```

```python
counter_lists = dict(Counter(results_string.replace('\n'," ").split()))
print(counter_lists)
```

```python
df = pd.DataFrame(data=counter_lists, index=[0])

df = (df.T)

print (df)

df.to_excel('Word Frequency.xlsx')
```
